# Laporan Hasil Training: MultiNLI

## Ringkasan Eksekutif
Model dilatih untuk menyelesaikan tugas *Natural Language Inference* (NLI), yaitu menentukan hubungan logika (Entailment, Neutral, Contradiction) antara dua kalimat (*Premise* & *Hypothesis*).

## Metrik Evaluasi
Menggunakan split validasi `validation_matched` dengan hasil sebagai berikut:

| Epoch | Training Loss | Validation Loss | Accuracy |
|-------|---------------|-----------------|----------|
| 1     | No log        | 0.9012          | 59.70 %  |
| 2     | No log        | 0.8074          | 64.30 %  |
| 3     | No log        | 0.8018          | 65.50 %  |

*(Data diambil dari output training notebook Task 3)*

## Analisis
- **Pemahaman Konteks:** Akurasi meningkat secara bertahap dari 59.7% ke 65.5%, menunjukkan model mulai mempelajari pola logika antar kalimat meskipun dilatih dengan data terbatas.
- **Keterbatasan Log:** Kolom *Training Loss* bernilai "No Log" kemungkinan karena parameter `logging_steps` pada `TrainingArguments` diatur lebih besar dari jumlah step per epoch (471 steps), sehingga log training loss tidak tercatat di setiap epoch.
- **Limitasi:** Akurasi ~65% tergolong moderat. Mengingat tugas NLI lebih kompleks daripada klasifikasi teks biasa, penggunaan dataset penuh (bukan subset 5000 sampel) sangat disarankan untuk mencapai performa yang lebih optimal (biasanya >80%).
